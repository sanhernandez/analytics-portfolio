```python
%%sql

CREATE TABLE campaign_data (
    user_id INTEGER,
    group_name TEXT,
    impressions INTEGER,
    clicks INTEGER,
    conversions INTEGER,
    revenue INTEGER
);
```

```python
%config SqlMagic.style = '_DEPRECATED_DEFAULT'
```

```python
%%sql

SELECT * FROM campaign_data;
```

```python
%%sql

SELECT 
    group_name,
    SUM(conversions) AS total_conversions,
    SUM(clicks) AS total_clicks,
    SUM(conversions) * 1.0 / SUM(clicks) AS conversion_rate
FROM campaign_data
GROUP BY group_name;
```

```python
%%sql

SELECT 
    group_name,
    SUM(clicks) * 1.0 / SUM(impressions) AS ctr
FROM campaign_data
GROUP BY group_name;
```

```python
%%sql

SELECT 
    group_name,
    SUM(revenue) AS total_revenue
FROM campaign_data
GROUP BY group_name;
```
